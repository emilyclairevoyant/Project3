```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Take in all of our WQI data and read it into pandas
WQI_2018 = "Resources/2018_data.csv"
WQI_2019 = "Resources/2019_data.csv"
WQI_2020 = "Resources/2020_data.csv"
WQI_2021 = "Resources/2021_data.csv"
WQI_2022 = "Resources/2022_data.csv"


WQI_2018_df = pd.read_csv(WQI_2018)
WQI_2019_df = pd.read_csv(WQI_2019)
WQI_2020_df = pd.read_csv(WQI_2020)
WQI_2021_df = pd.read_csv(WQI_2021)
WQI_2022_df = pd.read_csv(WQI_2022)
```

```python
# Rename columns to include the year
WQI_2018_df = WQI_2018_df.rename(columns=lambda x: f"{x}_2018")
WQI_2019_df = WQI_2019_df.rename(columns=lambda x: f"{x}_2019")
WQI_2020_df = WQI_2020_df.rename(columns=lambda x: f"{x}_2020")
WQI_2021_df = WQI_2021_df.rename(columns=lambda x: f"{x}_2021")
WQI_2022_df = WQI_2022_df.rename(columns=lambda x: f"{x}_2022")
```

```python
# Merge the DataFrames on Country Name column
consolidated_WQI_df = WQI_2018_df.merge(WQI_2019_df, left_on='Country Name_2018', right_on='Country Name_2019', how='outer') \
                                   .merge(WQI_2020_df, left_on='Country Name_2018', right_on='Country Name_2020', how='outer') \
                                   .merge(WQI_2021_df, left_on='Country Name_2018', right_on='Country Name_2021', how='outer') \
                                   .merge(WQI_2022_df, left_on='Country Name_2018', right_on='Country Name_2022', how='outer')
consolidated_WQI_df.head()
```

```python
# Rename the columns back to their original names
consolidated_WQI_df.rename(columns={
    'Country Name_2018': 'Country Name',
    'Country Code_2018': 'Country Code'},inplace = True)
consolidated_WQI_df.head()
```

```python
# Handle null values
consolidated_WQI_df.fillna('',inplace=True)
consolidated_WQI_df.head()
```

```python
#drop unnnecessary columns 
columns_to_drop = ['Country Name_2019', 'Country Code_2019','Country Name_2020', 'Country Code_2020', 'Country Name_2021', 'Country Code_2021', 'Country Name_2022', 'Country Code_2022']

# Drop the specified columns
# Drop the specified columns and assign the result to a new DataFrame
consolidated_WQI_data = consolidated_WQI_df.drop(columns=columns_to_drop, errors='ignore')
print(consolidated_WQI_data.head())
```

```python
# Save the consolidated DataFrame to a CSV file
consolidated_WQI_data.to_csv('Resources/consolidated_WQI_data.csv', index=False)
```

```python
# Take in all of our WQI data and read it into pandas
Population = "Resources\API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv"
Quality_of_Life = "Resources\standard-of-living-by-country-_-quality-of-life-by-country-2024.csv"

Population_df = pd.read_csv(Population)
Quality_of_Life_df = pd.read_csv(Quality_of_Life)

Population_df.head()
```

```python
#Replace the column name in the Quality_of_life_df
Quality_of_Life_df.rename(columns={
    'country': 'Country Name'
   },inplace = True)
Quality_of_Life_df.head()
```

```python
# Filter to keep only the specified columns for QoL Scores
QoL_columns = [
    'Country Name','StandardOfLiving_QoLScoreNumbeo_2023MidYear']
# Create the QoL Scores Data Frame
QoL_scores = Quality_of_Life_df[QoL_columns]
# create a new column with heading as Rank and find the Ranks based on the QoL score values 
# The highest quality index gets rank 1
QoL_scores.loc[:, 'Rank'] = QoL_scores['StandardOfLiving_QoLScoreNumbeo_2023MidYear'].rank(ascending=False)
#Replace the column name in the QoL_scores
QoL_scores.rename(columns={
    'StandardOfLiving_QoLScoreNumbeo_2023MidYear': 'Quality of Life (Numbeo 2023) '
   },inplace = True)
QoL_scores.head()
```

```python
# Filter to keep only the specified columns for population
Population_columns = [
    'Country Name','Country Code','2023']
# Create the Demographics Data Frame
Demographics_df = Population_df[Population_columns]
#Replace the column name in the Demographics_df
Demographics_df.rename(columns={
    '2023': 'Population'
   },inplace = True)
Demographics_df.head()
```

```python
# Merge the DataFrames on Country Name column, including only common country names
Summary_df = Demographics_df.merge(QoL_scores, on='Country Name', how='inner')
Summary_df.head()
```

```python
num_rows = len(Summary_df)  # This also gives you the number of rows
print("Number of rows in Summary_df:", num_rows)
```

```python
# Save the summary DataFrame to a CSV file
Summary_df.to_csv('Resources/Summary_Info.csv', index=False)
```
