```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import re
```

```python
# Take in all of our WQI data and read it into pandas
WQI_2018 = "Resources/2018_data.csv"
WQI_2019 = "Resources/2019_data.csv"
WQI_2020 = "Resources/2020_data.csv"
WQI_2021 = "Resources/2021_data.csv"
WQI_2022 = "Resources/2022_data.csv"


WQI_2018_df = pd.read_csv(WQI_2018)
WQI_2019_df = pd.read_csv(WQI_2019)
WQI_2020_df = pd.read_csv(WQI_2020)
WQI_2021_df = pd.read_csv(WQI_2021)
WQI_2022_df = pd.read_csv(WQI_2022)
```

```python
# Rename columns to include the year
WQI_2018_df = WQI_2018_df.rename(columns=lambda x: f"{x}_2018")
WQI_2019_df = WQI_2019_df.rename(columns=lambda x: f"{x}_2019")
WQI_2020_df = WQI_2020_df.rename(columns=lambda x: f"{x}_2020")
WQI_2021_df = WQI_2021_df.rename(columns=lambda x: f"{x}_2021")
WQI_2022_df = WQI_2022_df.rename(columns=lambda x: f"{x}_2022")
```

```python
# Merge the DataFrames on Country Name column
consolidated_WQI_df = WQI_2018_df.merge(WQI_2019_df, left_on='Country Name_2018', right_on='Country Name_2019', how='outer') \
                                   .merge(WQI_2020_df, left_on='Country Name_2018', right_on='Country Name_2020', how='outer') \
                                   .merge(WQI_2021_df, left_on='Country Name_2018', right_on='Country Name_2021', how='outer') \
                                   .merge(WQI_2022_df, left_on='Country Name_2018', right_on='Country Name_2022', how='outer')
consolidated_WQI_df.head()
```

```python
# Rename the columns back to their original names
consolidated_WQI_df.rename(columns={
    'Country Name_2018': 'Country Name',
    'Country Code_2018': 'Country Code'},inplace = True)
consolidated_WQI_df.head()
```

```python
# Handle null values
consolidated_WQI_df.fillna('',inplace=True)
consolidated_WQI_df.head()
```

```python
#drop unnnecessary columns 
columns_to_drop = ['Country Name_2019', 'Country Code_2019','Country Name_2020', 'Country Code_2020', 'Country Name_2021', 'Country Code_2021', 'Country Name_2022', 'Country Code_2022']

# Drop the specified columns
# Drop the specified columns and assign the result to a new DataFrame
consolidated_WQI_data = consolidated_WQI_df.drop(columns=columns_to_drop, errors='ignore')
print(consolidated_WQI_data.head())
```

```python
# Display all column names in the consolidated_WQI_data DataFrame
print(consolidated_WQI_data.columns.tolist())
```

```python
p = [r'Country Name', r'Country Code', r'Mobile' , r'electricity', r'Internet', r'SPI']
infrastructure_list = []
for i in p:
    infrastructure_list.append(consolidated_WQI_data.filter(regex = i))
infrastructure_df = pd.merge(infrastructure_list[0],infrastructure_list[1], how='outer', left_index=True, right_index=True)
```

```python
infrastructure_df = infrastructure_df.join([infrastructure_list[2], infrastructure_list[3], infrastructure_list[4], infrastructure_list[5]])
```
