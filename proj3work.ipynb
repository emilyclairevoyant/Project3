```python
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client['proj3test']
```

```python
dict_2018= data_2018.to_dict('records')
dict_2019= data_2019.to_dict('records')
dict_2020= data_2020.to_dict('records')
dict_2021= data_2021.to_dict('records')
dict_2022= data_2022.to_dict('records')
```

```python
for record in dict_2018:
    if '_id' in record:
        del record['_id']

db['collection_2018'].insert_many(dict_2018)
db['collection_2019'].insert_many(dict_2019)
db['collection_2020'].insert_many(dict_2020)
db['collection_2021'].insert_many(dict_2021)
db['collection_2022'].insert_many(dict_2022)
```
